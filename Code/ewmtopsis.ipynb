```python
#bihar
import pandas as pd 
import numpy as np
import seaborn as sns
df = pd.read_csv('finalcompileddatabihar.csv')
print("Initial DataFrame:")
print(df.head())

# Decision Matrix
df.index = df['Year']
new_df = df.drop([ 'Year' ], axis = 1)
print("\nDecision Matrix:")
print(new_df.head())


#Normalize Decision matrix

def norm(X):
    return X/X.sum()


norm_df = new_df.apply(norm)
print("\nNormalized Decision Matrix:")
print(norm_df.head())

#Entropy Values

k = -(1/np.log(norm_df.shape[0]))

def entropy(X):
    return (X*np.log(X)).sum()*k

entropy = norm_df.apply(entropy)
print("\nNormalized Decision Matrix:")
print(norm_df.head())

#degree of differentiation

dod = 1 - entropy

w = dod/dod.sum()
print("\nWeights (Degree of Differentiation):")
print(w.sort_values(ascending=False))

#TOPSIS
new_df.head()
def norm(X):
    return X/np.sqrt((X**2).sum())

norm_matrix = new_df.apply(norm)
norm_matrix.head()
w_norm_matrix = norm_matrix*w
w_norm_matrix.head()
V_plus = w_norm_matrix.apply(max)
V_minus = w_norm_matrix.apply(min)
V_plus
S_plus = np.sqrt(((w_norm_matrix - V_plus)**2).apply(sum, axis = 1))
S_minus = np.sqrt(((w_norm_matrix - V_minus)**2).apply(sum, axis = 1))
p_score = S_minus/(S_plus + S_minus)
print("\nTOPSIS Scores:")
print(p_score.sort_values(ascending=False).head(20))
```
